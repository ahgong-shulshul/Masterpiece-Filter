### 가장 최신 파일 경로 가져오기

```python
import os
from pathlib import Path

def get_latest_image_path(directory_path):
    # 디렉토리에서 이미지 파일을 찾기
    image_files = [f for f in os.listdir(directory_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    if not image_files:
        print("디렉토리에 이미지 파일이 없습니다.")
        return None

    # 이미지 파일들의 수정 시간 확인하여 가장 최신 파일 찾기
    latest_image = max(image_files, key=lambda f: os.path.getmtime(os.path.join(directory_path, f)))
    latest_image_path = os.path.join(directory_path, latest_image)

    return latest_image_path

# 테스트를 위한 디렉토리 경로 (적절한 디렉토리 경로로 교체하세요)
directory_path = '/content/drive/MyDrive/fast-style-transfer/app/examples/content/user-id-1'

latest_image_path = get_latest_image_path(directory_path)

if latest_image_path:
    print(f'가장 최신 이미지의 경로: {latest_image_path}')
else:
    print('이미지를 찾을 수 없습니다.')
```

### url 이미지 로컬에 저장하기

```python
import requests
from urllib.parse import urlparse

def download_image_url(url, user_id, image_name):
    # URL에서 파일 이름 추출
    parsed_url = urlparse(url)
    file_name = os.path.basename(parsed_url.path)

    # 파일 확장자 추출
    _, file_extension = os.path.splitext(file_name)
    file_extension = file_extension.lower()

    # 이미지 다운로드
    response = requests.get(url)
    if response.status_code == 200:
        # 로컬에 저장할 파일 경로 결정
        save_file_path = f'/content/{image_name}'
        print("save_file_path: ", save_file_path)
        # 이미지를 바이너리 모드로 저장
        with open(save_file_path, 'wb') as file:
            file.write(response.content)

        print(f'이미지 다운로드 성공. 저장 경로: {save_file_path}')
    else:
        print(f'이미지 다운로드 실패. 상태 코드: {response.status_code}')

download_image_url('https://sooking87.github.io/assets/image/ML/forest.jpg', '1', 'sibal.jpg')
```
